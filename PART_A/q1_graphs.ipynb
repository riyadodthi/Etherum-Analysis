```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime as dt
```

### Create a bar plot showing the number of transactions occurring every month between the start and end of the dataset.

```python

with open("Output/q1_count.csv") as q1:
    mylist = q1.read()
  

mylist=mylist.replace('[', '').replace(']', '').replace('"','').replace(' ','')

my_list=mylist.split(",")


date= [dt.strptime(my_list[i], '%m-%Y') for i in range(len(my_list)) if i%2==0] 
number =[int(my_list[i]) for i in range(len(my_list)) if i%2==1]


df = pd.DataFrame(data={'date':date,'number':number}).sort_values(['date'])
display(df.tail())
total_count=sum(df['number'][:])
print(total_count)
plt.figure(figsize=(10,5))
bar = plt.bar(df['date'],df['number'],width=15)
plt.title("No. of transactions between 08-2015 to 01-2019")
plt.ylabel("Number of transactions (le7)")
plt.xlabel("Year-Month")
#plt.xlim(min(df['date']), max(df['date']))
plt.xticks(rotation = 90, fontsize=7) 

plt.show()
```

### Create a bar plot showing the average value of transaction in each month between the start and end of the dataset.

```python
with open("Output/q1_avg.csv") as q1:
    mylist = q1.read()
  
mylist=mylist.replace('[', '').replace(']', '').replace('"','').replace(' ','')

my_list=mylist.split(",")
#print(my_list)

date= [dt.strptime(my_list[i], '%m-%Y') for i in range(len(my_list)) if i%2==0] 
number =[float(my_list[i]) for i in range(len(my_list)) if i%2==1]


df = pd.DataFrame(data={'date':date,'number':number}).sort_values(['date'])
display(df.head())
plt.figure(figsize=(10,5))
bar = plt.bar(df['date'],df['number'],width=15)
plt.title("Avg Transactions b/w 08-2015 to 01-2019")
plt.ylabel("Avg transactions (le20)")
plt.xlabel("Year-Month")

plt.xticks(rotation = 90, fontsize=7) 
plt.show()

    
```
