```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime as dt
```

```python

with open("Final_O/q4gasprice.txt") as q4:
    mylist = q4.read()
  

mylist = mylist.replace('[', '').replace(']', '').replace('"','').replace(' ','').replace('(','').replace(')','').replace('\'','')

my_list=mylist.split(",")
my_list_new = [val for val in my_list if val != 'Scamming']
#print(my_list_new)

#date= [dt.strptime(my_list_new[i], '%m-%Y') for i in range(len(my_list_new)) if i%2==0]
date= ['{:%Y-%m}'.format(dt.strptime(my_list_new[i], '%m-%Y')) for i in range(len(my_list_new)) if i%2==0] 
number =[float(my_list_new[i]) for i in range(len(my_list_new)) if i%2==1]

#print(date)
#print(number)


df = pd.DataFrame(data={'date':date,'number':number}).sort_values(['date'])
display(df.tail())

plt.figure(figsize=(20,6))
plot = plt.plot(df['date'],df['number'])
plt.title("Average gas price change over time")
plt.ylabel("Price")
plt.xlabel("Year-Month")
plt.xticks(rotation = 90, fontsize=9) 
plt.show()
```
